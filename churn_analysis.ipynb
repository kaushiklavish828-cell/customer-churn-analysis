```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded!")
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
print(df.shape)
df.head()
```

```python
print(df['Churn'].value_counts())
print("\nChurn Rate:", round(df['Churn'].value_counts(normalize=True)['Yes']*100, 2), "%")
```

```python
df['Churn'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen','salmon'])
plt.title('Customer Churn Distribution')
plt.ylabel('')
plt.savefig('churn_pie.png')
plt.show()
```

```python
sns.countplot(data=df, x='Contract', hue='Churn', palette='Set2')
plt.title('Contract Type vs Churn')
plt.savefig('contract_churn.png')
plt.show()
```

```python
sns.boxplot(data=df, x='Churn', y='tenure', palette='Set2')
plt.title('Tenure vs Churn')
plt.savefig('tenure_churn.png')
plt.show()
```

```python
print("=== KEY INSIGHTS ===")
print("1. Churn Rate:", round(df['Churn'].value_counts(normalize=True)['Yes']*100,2), "%")
print("2. Avg tenure (Churned):", round(df[df['Churn']=='Yes']['tenure'].mean(),1), "months")
print("3. Avg tenure (Stayed):", round(df[df['Churn']=='No']['tenure'].mean(),1), "months")
```
